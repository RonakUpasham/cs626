```python
import os
import pandas as pd
from collections import defaultdict
```

```python
# Define the paths to your positive and negative data folders
origin_path = "D:/OneDrive - Indian Institute of Technology Bombay/Study/Semester 5/CS 626/Project/review_polarity/"
positive_folder = origin_path + "txt_sentoken/pos"
negative_folder = origin_path + "txt_sentoken/neg"
```

```python
# Initialize empty lists to store data
reviews = []
labels = []

# Read positive sentiment reviews
for filename in os.listdir(positive_folder):
    with open(os.path.join(positive_folder, filename), 'r', encoding='utf-8') as file:
        review = file.read()
        reviews.append(review)
        labels.append('positive')

# Read negative sentiment reviews
for filename in os.listdir(negative_folder):
    with open(os.path.join(negative_folder, filename), 'r', encoding='utf-8') as file:
        review = file.read()
        reviews.append(review)
        labels.append('negative')

# Create a DataFrame
panglee = pd.DataFrame({'Review': reviews, 'Label': labels})
panglee[["len"]]  = panglee[["Review"]].applymap(len)
panglee.to_csv("./data/panglee_sup.csv")
panglee
```

```python
# Your list of numbers
panglee_len = list(panglee["len"])

# Define the range size (e.g., 1000)
range_size = 1000

# Create a defaultdict to count frequencies in ranges
frequency_distribution = defaultdict(int)

# Count frequencies in ranges
for number in panglee_len:
    range_start = (number // range_size) * range_size
    range_end = range_start + range_size
    range_label = f"{range_start}-{range_end}"
    frequency_distribution[range_label] += 1

# Display the frequency distribution
for range_label, frequency in frequency_distribution.items():
    print(f"Range: {range_label}, Frequency: {frequency}")
```
